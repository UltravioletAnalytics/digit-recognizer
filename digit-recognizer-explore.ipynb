```python
import pandas as pd
pd.__version__
```

```python
import h2o
from h2o.estimators.deeplearning import H2ODeepLearningEstimator

h2o.init()
```

```python
# For some reason H2O can't handle relative paths(?), so specify entire path
train_df = h2o.import_file('/home/dlnovell/projects/kaggle/digit-recognizer/data/train.csv')
train_df.describe()
```

```python
test_df = h2o.import_file('/home/dlnovell/projects/kaggle/digit-recognizer/data/test.csv')
test_df.describe()
```

```python
# There are an awful lot of pixels that have a max value of 0 - so apparently they are never set. Perhaps we can
# improve or at least speed up training by removing any pixels that are never set. If any test example of a character
# has a pixel set that's never been encountered before, it may contribute noise to the model. Likely a small effect,
# but worth experimenting with
```

```python
# Convert the datatype of the label column from integer to factor 
train_df['label'] = train_df['label'].asfactor()
```

```python
# Create a validation set
r = train_df.runif()
train_split = train_df[r < 0.8]
valid_split = train_df[r >= 0.8]
```

```python
# Create the model object with some reasonable parameters
model = H2ODeepLearningEstimator(
            distribution="multinomial",
            activation="RectifierWithDropout",
            hidden=[100,200,100],
            input_dropout_ratio=0.2,
            sparse=True,
            l1=1e-5,
            epochs=100)

model.params

```

```python
# Train the model on the training data, specifying the validation data
model.train(
        x=train_df.names[1:785],
        y=train_df.names[0],
        training_frame=train_split,
        validation_frame=valid_split)
```

```python
# Take a look at the results of the model training
model
```

```python
# Using the test dataframe, make predictions using the trained model
pred = model.predict(test_df)
```

```python
# Examine what was predicted for the top 10 rows
pred.head()
```

```python
# Prepare the submission for Kaggle by generating an "ImageID" column and appending the prediction label
submission_df = h2o.H2OFrame({'ImageID': list(range(1, len(test_df)+1))}).cbind(pred[:,0])
submission_df.set_name(1, "Label")
```

```python
# Save the submission with a timestamp
from datetime import datetime
timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
print timestamp
h2o.export_file(submission_df, path='/home/dlnovell/projects/kaggle/digit-recognizer/submissions/submission_' + timestamp + '.csv')
```
